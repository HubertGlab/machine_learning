# Object detection

# Importacja plików i ich przygotowanie

```python
# Biblioteki
import cv2 # pip install opencv-python 
import matplotlib.pyplot as plt # pip install matplotlib
import math
import numpy as np
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file) 
```

```python
# odczytanie dokumentu z labelsami (do oczytu z obrazków/filmów)
classLabels = []
file_name = 'names.txt'
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
    
print(classLabels)
```

```python
# ustawienia modelu
model.setInputSize(320, 320)
model.setInputScale((1.0/127.5)) 
model.setInputMean((127.5, 127.5, 127.5))
model.setInputSwapRB(True)
```

# Odczytanie obrazka

```python
img = cv2.imread('personandcar.png') # Wczytanie obrazka
```

```python
plt.imshow(img) # Pokazanie obrazka
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # Zmiana z bgr na rgb
```

```python
ClassIndex, confidece, bbox = model.detect(img,confThreshold=0.5) 
# zmienne w tym zmiennna confThreshold którą zmniejszając uzyskamy więcej odczytów z obrazka, 
# lecz mogą być one mniej poprawne, i adekwatnie zwiększając uzyskamy mniej odczytów ale bardziej pewne
```

```python
print(ClassIndex) # Odczytanie wyników (zgodnie z plikiem labels 1 to person a 3 to car)
```

```python
# wyswietlanie 
font_scale = 2 # rozmiar czcionki napisów
font = cv2.FONT_HERSHEY_PLAIN  # czcionka
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
    cv2.rectangle(img, boxes, (255, 0, 0), 2) 
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness = 2)
    
    # dodanie otoczek i podpisów do obrazku, do rozpoznanych przez program obiektów
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) # pokazanie obrazka, już wraz z oznaczeniem obiektów, oraz zmiana z bgr na rgb
```

# Odczytanie filmiku

```python
cap = cv2.VideoCapture("Pexels Videos 2880.mp4")

if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError("Cannot open video")
    
    # załadowanie filmiku i sprawdzenie czy się otwiera
    
font_scale = 2
font = cv2.FONT_HERSHEY_PLAIN 

while True:
    ret,frame = cap.read()
    
    ClassIndex, confidece, bbox = model.detect(frame, confThreshold = 0.40)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidece.flatten(), bbox):
            if (ClassInd <= 80):
                cv2.rectangle(frame, boxes, (255, 0 , 0), 2)
                cv2.putText(frame, classLabels[ClassInd-1],(boxes[0]+10,boxes[1]+40), font, fontScale=font_scale, color=(0, 255, 0), thickness = 2)
                
    cv2.imshow('Wykrywacz', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```

# Object tracking

```python
import cv2
cap = cv2.VideoCapture("yoo4.mp4") # kamerka

tracker = cv2.TrackerCSRT_create() # tracker myszki
success, img = cap.read() 
bbox = cv2.selectROI("Tracking", img, )
tracker.init(img, bbox)

def drawBox(img, bbox): 
    x, y, w, h = int(bbox[0]), int(bbox[1]), int(bbox[2]), int(bbox[3])
    cv2.rectangle(img, (x, y), ((x+w), (y+h)), (255,0,255),3,1) # rysowanie ramki. bbox zawiera parametry x y w i h, 
    # img pojedyńcza klatka
    
while True: # klatka po klatce
    success, img = cap.read() 
    success, bbox = tracker.update(img)
    print(type(bbox))
    if success:
        drawBox(img, bbox)
    
    cv2.imshow("Tracking", img)
    
    if cv2.waitKey(1) & 0xff == ord('q'): # wyłączenie
        break 
        
cap.release()
cv2.destroyAllWindows()
```
